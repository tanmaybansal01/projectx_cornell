```python
rides_a = pd.read_csv('../data/Rides_DataA.csv')
rides_b = pd.read_csv('../data/Rides_DataB.csv')
rides = pd.merge(rides_a, rides_b, on='RIDE_ID')
rides["started_on"] = pd.to_datetime(rides["started_on"], utc=True)
rides["started_on_date"] = rides["started_on"].dt.date
rides["started_on_hour"] = rides["started_on"].dt.hour
rides["started_on_day"] = rides["started_on"].dt.dayofweek
```

```python
rides = rides[rides["end_location_lat"] > 28]
rides = rides[rides["end_location_lat"] < 35] 
rides = rides[rides["end_location_long"] < -92]
```

```python
demand_df = pd.read_csv("../ride_predictions_updated_2/hourly_predictions_0.csv")[[
        "started_on_hour", "stacked_pred"]]
col = "predicted_demand_0"
demand_df = demand_df.rename(columns={"stacked_pred": col})
demand_df[col] = demand_df[col].interpolate()
for i in range(1, 10): 
    col = "predicted_demand_" + str(i)
    df = pd.read_csv("ride_predictions_updated/hourly_predictions_" + str(i) + ".csv")[[
        "started_on_hour", "lstm_pred"]]
    df = df.rename(columns={"lstm_pred": col})
    df[col] = df[col].interpolate()
    demand_df = pd.merge(demand_df, df, on=["started_on_hour"])
demand_df["started_on_hour"] = pd.to_datetime(demand_df["started_on_hour"])
```

```python
demand_df["started_on_date"] = demand_df["started_on_hour"].dt.date
demand_df["started_on_hour"] = demand_df["started_on_hour"].dt.hour
def get_total_demand(row): 
    sum = 0 
    for i in range(10): 
        sum += row["predicted_demand_" + str(i)]
    return sum
demand_df["total_demand"] = demand_df.apply(get_total_demand, axis=1) 
```

```python
rides_with_demands = pd.merge(rides, demand_df, on=["started_on_date", "started_on_hour"])
```

```python
def haversine_distance(lat1, long1, lat2, long2, earth_radius=6371): # 6371 is radius in km
    lat1, long1, lat2, long2 = np.radians([lat1, long1, lat2, long2])
    a = np.sin((lat2-lat1)/2.0)**2 + \
        np.cos(lat1) * np.cos(lat2) * np.sin((long2-long1)/2.0)**2
    
    return earth_radius * 2 * np.arcsin(np.sqrt(a))
```

```python
centroids = pd.read_csv("../cluster_centers.csv")
```

```python
alpha = 0.6
```

```python
def deadheading_distance_predicted(arr): 
    arr["shifted_end_location_lat"] = arr["end_location_lat"].shift()
    arr["shifted_end_location_long"] = arr["end_location_long"].shift()
    def calculate_optimal_centroid(row):
        optimal_centroid = -1 
        max_v = -1
        max_dst = -1
        for zone in range(10): 
            distance_from_previous_ride = haversine_distance(row["shifted_end_location_lat"], row["shifted_end_location_long"], 
                                      centroids.iloc[zone]["0"], centroids.iloc[zone]["1"]) 
            dist = 100 / distance_from_previous_ride
            demand = (row["predicted_demand_" + str(zone)] / row["total_demand"])
            v = alpha * dist + (1 - alpha) * demand 
            if v > max_v: 
                optimal_centroid = zone 
                max_v = v
                max_dst = distance_from_previous_ride
        lat = centroids.iloc[optimal_centroid]["0"]
        lng = centroids.iloc[optimal_centroid]["1"]
        return (lat, lng, max_dst)
                
    def calculate_distance(row):
        centroid = calculate_optimal_centroid(row)
        driving_distance_to_rider = haversine_distance(centroid[0], centroid[1], 
                                      row["start_location_lat"], row["start_location_long"]) * 1.4 # network distance
        distance_from_previous_ride = centroid[2] * 1.4
        return driving_distance_to_rider + distance_from_previous_ride
    return arr.apply(calculate_distance, axis=1).sum()
```

```python
rides_grouped_actual = pd.read_csv("../actual_per_ride_deadheading.csv")
rides_grouped_actual["started_on_date"] = pd.to_datetime(rides_grouped_actual["started_on_date"])
rides_grouped_actual["started_on_date"] = rides_grouped_actual["started_on_date"].dt.date
```

```python
rides_grouped_predicted = rides_with_demands.groupby(["driver_id", "started_on_date"]).apply(deadheading_distance_predicted)
rides_grouped_predicted = rides_grouped_predicted.reset_index().rename(columns={0: "predicted_deadheading_distance"})
```

```python
rides_grouped = pd.merge(rides_grouped_actual, rides_grouped_predicted, on=["driver_id", "started_on_date"])
rides_grouped["difference"] = rides_grouped["ride_deadheading"] - rides_grouped["predicted_deadheading_distance"]
```

```python
rides_grouped = rides_grouped[rides_grouped["ride_deadheading"] > 0]
rides_grouped = rides_grouped[rides_grouped["predicted_deadheading_distance"] > 0]
```

```python
rides_grouped = rides_grouped[rides_grouped["difference"].between(rides_grouped[
    "difference"].quantile(.027), rides_grouped["difference"].quantile(.973))]
```

```python
emissions_df = pd.read_csv("../data/epa_emissions_database.csv")
rides_b = pd.read_csv('../data/Rides_DataB.csv')
rides_b["make"] = rides_b["make"].str.strip("b\'\"")
rides_b["model"] = rides_b["model"].str.strip("b\'\"")
rides_b["year"] = rides_b["year"].str.strip("b\'\"")
rides_b["year"] = pd.to_numeric(rides_b["year"])
emissions_df = emissions_df.rename(columns={"Make": "make", "Model": "model", "Year": "year", 
                                            "Tailpipe CO2 in Grams/Mile (FT1)": "emissions"})
emissions_df = emissions_df.groupby(["make", "model", "year"])["emissions"].median()
emissions_df = pd.DataFrame(emissions_df)
emissions_df = emissions_df[emissions_df["emissions"] >= 0]
emissions_df_merged = pd.merge(rides_b, emissions_df, on=["make", "model", "year"], how = "left")
emissions_df_merged = emissions_df_merged[["driver_id", "make", "model", "year", "emissions"]]
emissions_df_merged = emissions_df_merged.drop_duplicates()
emissions_df_merged["emissions"] = emissions_df_merged["emissions"].fillna(emissions_df_merged["emissions"].median())
```

```python
rides_with_emissions = pd.merge(rides_grouped, emissions_df_merged, on=["driver_id"])
```

```python
rides_with_emissions["actual_emissions_in_g"] = rides_with_emissions["emissions"] * rides_with_emissions["ride_deadheading"] * 0.621371
rides_with_emissions["predicted_emissions_in_g"] = rides_with_emissions["emissions"] * rides_with_emissions["predicted_deadheading_distance"] * 0.621371
```
